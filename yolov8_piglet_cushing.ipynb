```python
%cd /
from glob import glob
img_list = glob('/content/gdrive/My Drive/yolov8_piglet_crushing/dataset/export/images/*.jpg')
print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split
train_img_list, val_img_list = train_test_split(img_list, test_size = 0.3, random_state = 1000)
print(len(train_img_list), len(val_img_list))
```

```python
with open('/mydrive/yolov8_piglet_crushing/dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/mydrive/yolov8_piglet_crushing/dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

data = {
    'train': '/mydrive/yolov8_piglet_crushing/dataset/train.txt',
    'val': '/mydrive/yolov8_piglet_crushing/dataset/val.txt',
    'test': '/mydrive/yolov8_piglet_crushing/dataset/test.txt',
    'nc': 3,
    'names': ['piglet', 'crushing', 'crushed']
}

with open('/mydrive/yolov8_piglet_crushing/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

with open('/mydrive/yolov8_piglet_crushing/dataset/data.yaml', 'r') as f:
  data_yaml = yaml.safe_load(f)
  display(data_yaml)
```

# Load a pre-trained model

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

# YOLOv8 커스텀 데이터 학습하기

```python
model.train(data='/mydrive/yolov8_piglet_crushing/dataset/data.yaml', epochs=10, patience=10, batch=16, imgsz=416)
```

```python
print(type(model.names), len(model.names))

print(model.names)
```
